```python
from bs4 import BeautifulSoup
import requests as r
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
'''text mining with using beautifulsoup and saving the text file with USER_ID number'''

df = pd.read_excel('data.xlsx')
urls = list(df['URL'])
url_id = list(df['URL_ID'])
a = list(zip(urls,url_id)) 

def scrap(url,file_name):
    url = url
    html_content = r.get(url).text
    soup = BeautifulSoup(html_content,'html.parser')
    file = open(file_name,"w" ,encoding="utf-8")
    tag = soup.find('title')
    file.write(tag.text)

    content = soup.find_all('div',class_= "td-post-content tagdiv-type")
    for i in content:
        file.write(i.text)
    
for i,ii in a:
    file_name = str(ii) + '.txt'
    url = i
    scrap(url,file_name)
    
```

```python

'''removing stop words from the text file, used nltk to perform sentiment analysis'''

#function to make data frame
def dataframe(sentiment_scorelist,df):
    new_data = pd.DataFrame(sentiment_scorelist)
    df = pd.concat([df,new_data],ignore_index=True)
    return df
    


#function for sentiment analysis
def sentiment_analysis(list1):
    list1 = list1
    analyse = SentimentIntensityAnalyzer()
    sentiment_score = analyse.polarity_scores(list1)
    return sentiment_score




#function for tokanization of scrap file
def tokenize_scrap(scrap_file):
    from nltk.tokenize import word_tokenize
    
  
    scrap_file = scrap_file
    File_Name = scrap_file + ".txt"
    file = open(File_Name,'r',encoding='latin-1')
    modified_text = file.read()
    upper_text = modified_text.upper()
    list1 = word_tokenize(upper_text)
    return list1



#function for tokanization of stop word file
def tokenize_StopWords(File_Name):
    from nltk.tokenize import word_tokenize
    File_Name = File_Name
    file = open(File_Name,"r")
    var = file.read()
    modified_text = var.upper()
    list2 = word_tokenize(modified_text)

    return list2
    
#function for removing stop words
def remove(list1,list2):
    list1 = list1
    list2 = list2
    list1 = list(filter(lambda x: x not in list2, list1))
    return list1

def List_words(List_File,list1):
    List_File = List_File
    list1 = list1
    for File_Name in List_File:
        list2 = tokenize_StopWords(File_Name)
        list1 = remove(list1,list2)
    return list1
    


#removing words
def removing_words(text_file):
    list1 = tokenize_scrap(text_file)
    List_File = ["StopWords_Auditor.txt","StopWords_Currencies.txt","StopWords_DatesandNumbers.txt","StopWords_Generic.txt","StopWords_GenericLong.txt","StopWords_Geographic.txt","StopWords_Names.txt"]
    list1 = List_words(List_File,list1)
    return list1



def text_file():
    a = 1
    df = pd.DataFrame({'neg':0.00,'neu':0.00,'pos':0.00, 'comp':0.00}, index=[a])
    sentiment_scorelist = []

    for text_file in range(39,150):
        text_file = str(text_file)
        list1 = removing_words(text_file)
        list1 = ", ".join(list1)
        sentiment_score = sentiment_analysis(list1)
        sentiment_scorelist.append(sentiment_score)
    output = dataframe(sentiment_scorelist,df)
    output.to_csv("output.csv")
    
        
    
text_file()    
```
